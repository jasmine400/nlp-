```python
df = pd.read_csv('astan_real(1).csv')
df = df.dropna()
print(type(df))
text=df['desc'].values
print(text)
```

```python
bad_chars = [';', ':', '!', "'","[",",","،","\n","؟",".",'"',"]",'🙏','✨','🌷','🌹','😘','🕌','☺','♀️','👏','۱۸۵۸','🙇','\u200d','🍃','😊','😥','❤','🤔','😔','😭','🙁','👌','😍','👍','👐','💙','😐','✅','🌺','🥀','😢','😞','😏','🍁','📙','😃','\u200c','✍']
a=[]
for item in text:
  b=str(item)
  for i in bad_chars :
   b = b.replace(i, '')
  a.append(b)
print(a)
```

```python
v=[]
for i in range(len(a)):
  s=str(a[i])
  s=s.split()
  v.append(s)
#print(v)
```

```python
data= open('dictio(1).txt').read()

counter = 0
wordList = list()

for group in data.split('\n'):
        for word in group.split('\t'):
                if (counter % 3 == 0):
                        wordList.append(word)
                counter += 1
a=["آ"]

for i in range((len(wordList))) :
  y=str(wordList[i])
  y=y.split("\u200c")
  a.extend(y)
print(a)
print(type(a))
res = [j for j in a if j not in wordList]
print(res)
```

```python
for j in v:
  i=1
  for i in range(len(j)-1):
    if j[i] in res:
      s1=j[i-1]+"\u200c"+j[i]
      s2=j[i]+j[i+1]
      for h in range(len(wordList)):
        if s1==wordList[h]:
          j[i-1]=s1
          j[i]=" "
        elif s2==wordList[h]:
          j[i]=s2
          j[i+1]=" "
```

```python
new_word= open('stopwords-fa.txt').read()
new_word=new_word.split()
lst=[]
for j in v:
  rem = [i for i in j if i not in new_word]
  lst.append(rem)
#print(lst)
```

```python

training_data, test_data = sklearn.model_selection.train_test_split(df, train_size = 0.7, random_state=42)
print(training_data.shape)
print(test_data.shape)
```

```python

bow_transform = CountVectorizer(tokenizer=lambda doc: doc, ngram_range=[3,3], lowercase=False)
```

```python

X_tr_bow = bow_transform.fit_transform(training_data['Text_Cleaned'])
```

```python
X_te_bow = bow_transform.transform(test_data['Text_Cleaned'])
```

```python
y_tr = training_data['real_label']
y_te = test_data['real_label']
```

```python
def simple_logistic_classify(X_tr, y_tr, X_test, y_test, description, _C=1.0):
    model = LogisticRegression(C=_C).fit(X_tr, y_tr)
    score = model.score(X_test, y_test)
    print('Test Score with', description, 'features', score)
    return model
```

```python
model_bow = simple_logistic_classify(X_tr_bow, y_tr, X_te_bow, y_te, 'bow')
```
